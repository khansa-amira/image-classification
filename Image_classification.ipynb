```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing import image

# Unzip the hewan dataset
!unzip -q /content/hewan.zip -d /content/

# Load dataset
train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/", # Changed path from "/content/fruits" to "/content/"
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(227,227),
    batch_size=32
)

val_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "/content/", # Changed path from "/content/fruits" to "/content/"
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(227,227),
    batch_size=50
)

# Ambil nama kelas otomatis
class_names = train_dataset.class_names

print("Class names:", class_names)

plt.figure(figsize=(10,8))

for images, labels in train_dataset.take(1):
    for i in range(6):
        ax = plt.subplot(2,3,i+1)

        plt.imshow(images[i].numpy().astype("uint8"))

        # label benar
        plt.title(class_names[int(labels[i])])

        plt.axis("off")

plt.show()


model = models.Sequential([

    # normalisasi pixel
    layers.Rescaling(1./255, input_shape=(227,227,3)),

    # CNN layer 1
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    # CNN layer 2
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    # CNN layer 3
    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(),

    # fully connected
    layers.Flatten(),
    layers.Dense(128, activation='relu'),

    # output (2 kelas: apple dan mango)
    layers.Dense(len(class_names), activation='softmax') # Dynamically set output classes
])

model.summary()
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=10
)
model.save("/content/model_hewan.keras")

print("Model berhasil disimpan")
```

```python
def prediksi_gambar(path_gambar):

    img = image.load_img(
        path_gambar,
        target_size=(227,227)
    )

    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    hasil = model.predict(img_array)

    prediksi = np.argmax(hasil)

    print("Prediksi :", class_names[prediksi])

    plt.imshow(img)
    plt.title(f"Hasil Prediksi: {class_names[prediksi]}")
    plt.axis("off")
    plt.show()


prediksi_gambar("/content/gambar_uji.jpg")
prediksi_gambar("/content/gambar_uji2.jpg")
```
