```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
import os # Import os module

# Define the model path
model_path = "/content/model_binatang.keras"

# Check if the file exists
if not os.path.exists(model_path):
    print(f"Error: The model file '{model_path}' was not found.")
    print("Please ensure the file is correctly uploaded and the path is accurate.")
else:
    try:
        # Load model hasil training
        model = tf.keras.models.load_model(model_path)
        print("Model berhasil dimuat")
    except ValueError as e:
        print(f"Error loading model: {e}")
        print("This often means the .keras file is corrupted or not in the expected format (e.g., not a valid zip archive).")
        print("Please verify the integrity of the file or try re-saving/re-uploading it.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
```

```python
class_names = ['ayam', 'gajah']
def prediksi_gambar(path_gambar):

    img = image.load_img(
        path_gambar,
        target_size=(227,227)
    )

    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)

    hasil = model.predict(img_array)

    prediksi = np.argmax(hasil)

    print("Probabilitas:", hasil)
    print("Prediksi:", class_names[prediksi])

    plt.imshow(img)
    plt.title(f"Hasil Prediksi: {class_names[prediksi]}")
    plt.axis("off")
    plt.show()
```

```python
prediksi_gambar("/content/gambar_uji.jpg")
prediksi_gambar("/content/gambar_uji2.jpg")
```
